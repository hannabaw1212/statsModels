# Point Estimate


## Properties of a Good Estimator


### Unbiased Estimator


```python
## assume that our population is distributed normal with mue=5 and sigma=3
## Let's generate samples and have our point estimators
## we want to estimate mue
estimators = []

for expiremnt in range(1000):
    data = np.random.normal(loc=5, scale=3, size=10000)
    estimator = np.mean(data)
    estimators.append(estimator)

print(f"E(estimatros) = {round(np.mean(estimators), 2)}")
print(f"E(estimatros) - true_val = {round(np.mean(estimators), 2) - 5} ")
```

### Consistency

```python
## assume that our population is distributed normal with mue=5 and sigma=3
## Let's generate samples and have our point estimators
## we want to estimate mue


expirments = 1000
n = [10, 50, 100, 200, 300, 500, 600, 700, 800, 900, 1000, 2000, 3000, 4000, 5000, 6000, 7000, 8000, 9000, 10000, 20000, 30000, 40000, 50000, 60000]  
thetas = []
for size in n:
    values = []
    for expiremnt in range(expirments):
        samples = np.random.normal(loc=5, scale=3, size=size)
        estimation = np.mean(samples)
        values.append(estimation)
    thetas.append(np.mean(values))
    
plt.plot(n, thetas)
plt.axhline(y=5, color='r', linestyle='-')
plt.xlabel("n")
plt.ylabel("theta")
```

### Newton – Raphson Method for MLE Calculation

```python
import numpy as np

def newton_raphson_method(theta):
    for iteration in range(5000):
        theta = theta - 0.001 * (Dfx(theta) / Dfx_2(theta))
    
    return theta

def f(x):
    return 2*x**4 + 5*x**2 + 5*x

def Dfx(x):
    return 8*x**3 + 10*x + 5

def Dfx_2(x):
    return 24*x**2 + 10

# Example usage
initial_guess = 0
result = newton_raphson_method(initial_guess)
print("Root approximation:", result)
```

### The Crude Monte Carlo Approach for Estimation

```python
## assume g is a normal distribution g ~ N(3, 5)

def crude_monte_carlo():
    g = sp.stats.norm(loc=3, scale=5)
    U = np.random.uniform(0, 1, size=1000)
    g_u = g.ppf(U)
    theta_CMC = np.mean(g_u)
    return theta_CMC

crude_monte_carlo()
```

## Variance Reduction


```python
crude_monte_carlo_results = np.array([crude_monte_carlo() for i in range(1000)])
var_cmc_l = np.var(crude_monte_carlo_results)
print(f"var(l_cmc)= {var_cmc_l}")
```

### Antithetic Variables


```python
import numpy as np
import scipy.stats as sp

def antithetic_variables():
    # Define the normal distribution N(3, 5)
    g = sp.norm(loc=3, scale=5)
    
    # Generate uniform random variables and their antithetic counterparts
    U = np.random.uniform(0, 1, size=1000)
    U_tag = 1 - U
    
    # Transform both U and U_tag to the normal distribution
    transformed_U = g.ppf(U)
    transformed_U_tag = g.ppf(U_tag)
    
    # Compute the average of the transformed variables
    theta_AV = np.mean((transformed_U + transformed_U_tag) / 2)
    
    print("Antithetic Variate Estimate:", theta_AV)
    return theta_AV

# Call the function to see the result
antithetic_estimate = antithetic_variables()
```
