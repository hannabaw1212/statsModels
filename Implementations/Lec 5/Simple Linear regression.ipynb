```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy as sp
```

# Simple Linear Regression


## The Optimal Linear Predictor

```python
x = np.linspace(-100, 100, 100)
y = 3*x + 5 + np.random.normal(0, 10, size=100)

sns.scatterplot(x=x, y=y)
plt.show()
```

```python
x = np.linspace(-100, 100, 100)
y = 3 * x + 5 + np.random.normal(0, 10, 100)  # Adding some noise

# Calculate the slope (b_1) and intercept (b_0) for the regression line
b_1 = np.cov(x, y)[0, 1] / np.var(x)
b_0 = np.mean(y) - b_1 * np.mean(x)

# Calculate the regression line values
y_pred = b_1 * x + b_0

# Create the plot
plt.figure(figsize=(10, 6))
plt.scatter(x, y, color='blue', label='Data points')  # Scatter plot of the data points
plt.plot(x, y_pred, color='red', label='Regression Line')  # Line plot for the regression line

# Add labels and title
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Linear Regression Plot')

# Show legend
plt.legend()

# Display the plot
plt.show()
```
